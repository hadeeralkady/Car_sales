<p style="background-color:#008000;font-family: 'Georgia';color:#FFFFFF;font-size:200%;text-align:center;border-radius:10px 10px;"> Data Analysis </p>

<span style="font-size:30px; font-family: 'Georgia', serif; color:#008000; font-weight:bold;"> Sales trends over time </span>

```python
sales_trends = df.groupby(df['Date'].dt.to_period('M')).size()
print("Sales Trends Over Time:\n", sales_trends)
```

```python
sales_trends = df.groupby(df['Date'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
plt.plot(sales_trends.index.to_timestamp(), sales_trends.values, marker='o', linestyle='-', color='blue', label='Sales')

plt.title('Sales Trends Over Time', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

<span style="font-size:30px; font-family: 'Georgia', serif; color:#008000; font-weight:bold;"> Price by date</span>

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

sales_trends = df.groupby(['Year', 'Month'])['Price ($)'].sum().reset_index()
sales_trends
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_trends, x='Month', y='Price ($)', hue='Year', marker='o')
plt.legend(title='Year')
plt.show()
```

<span style="font-size:30px; font-family: 'Georgia', serif; color:#008000; font-weight:bold;"> Top Companies</span>

```python
company_performance = df['Company'].value_counts()
print("\nTop Companies:\n", company_performance.head())
```

```python
plt.figure(figsize=(12, 8)) 

top_companies = company_performance.head(20) 
sns.barplot(x=top_companies.index, y=top_companies.values, palette="viridis")
plt.title('Top Companies by Sales', fontsize=16)
plt.xlabel('Company', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

<span style="font-size:30px; font-family: 'Georgia', serif; color:#008000; font-weight:bold;"> Performance by Model and Company </span>

```python
df['Model_Company'] = df['Model'] + ' - ' + df['Company']
company_model_counts = df['Model_Company'].value_counts()
print("\nTop Models per Company:\n", company_model_counts.head())
```

```python
plt.figure(figsize=(12, 8))  
top_models = company_model_counts.head(20) 
sns.barplot(x=top_models.index, y=top_models.values, palette="coolwarm")
plt.title('Top Models per Company by Sales', fontsize=16)
plt.xlabel('Model - Company', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.xticks(rotation=45) 
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

<span style="font-size:30px; font-family: 'Georgia', serif; color:#008000; font-weight:bold;"> Performance by Region</span>

```python
region_performance = df['Dealer_Region'].value_counts()
print("\nTop Sales per Regions:\n", region_performance.head())
```

```python
plt.figure(figsize=(12, 8)) 

top_regions = region_performance.head(20) 
sns.barplot(x=top_regions.index, y=top_regions.values, palette="magma")
plt.title('Top Regions by Sales', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

<span style="font-size:30px; font-family: 'Georgia', serif; color:#008000; font-weight:bold;"> sales by gender</span>

```python
sales_by_gender = df.groupby('Gender')['Price ($)'].sum().reset_index()
sales_by_gender
```

```python
plt.figure(figsize=(8, 5))
plt.bar(sales_by_gender['Gender'], sales_by_gender['Price ($)'])
plt.title('Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales ($)')
plt.show()
```

<span style="font-size:30px; font-family: 'Georgia', serif; color:#008000; font-weight:bold;"> sales by income brackets</span>

```python
income_brackets = pd.cut(df['Annual Income'], bins=[0, 30000, 60000, 100000, 150000, 200000, 300000], labels=['0-30K', '30K-60K', '60K-100K', '100K-150K', '150K-200K', '200K-300K'])
sales_by_income = df.groupby(income_brackets)['Price ($)'].sum().reset_index()
sales_by_income
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sales_by_income['Annual Income'], sales_by_income['Price ($)'])
plt.title('Sales by Annual Income')
plt.xlabel('Income Bracket')
plt.ylabel('Total Sales ($)')
plt.show()
```
